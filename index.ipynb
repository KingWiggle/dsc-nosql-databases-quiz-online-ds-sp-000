## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

The question of which style is better depends on the context in which they are applied.  SQL databases are table-based and excel at holding structured and predictable data.  NoSQL databases are not table-based and do not have a standard schema which makes them better at holding dynamic and unstructured data.  SQL databases support more complex querying while NoSQL databases may be better applied to large datasets for more efficient delegation of workloads.  


## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Storing social media data in a relational database would require an obscene amount of entities to reflect all the connections between profiles and groups.  NoSQL databases, graph databases in particular, make it much easier store this information using nodes (profiles) and edges (connections).  Depending on how active a user is, profile information can be small or large.  NoSQL databases handles this sort of dynamic information better than relational ones.  




## Question 3

How does MongoDB work? How is it different from a relational database?

MongoDB is an example of a document store database.  These databases store records as unique documents rather than in tables.  Each contains key-value pairs with the data store as the value.  It is better than tables at storing information on the fly but may be more complicated to query.




## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Fault tolerance has to do with making RDDs robust.  With the use of an increased number of servers to hold and process the data, there is an increased probability, at any point in time, that servers are not functioning properly.  Different blocks of data are stored on different servers and serve as backups in the event that a server is down.  Fault tolerance in RDDs allow the return of complete data even in less than optimal conditions.  




## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce is a processing technique and a program model for distributed computing (i.e. big data).  It is the technique behind RDDs and the engines that use them like Hadoop.  MapReduce refers to a two-step algorithm.  Map involves nodes, like servers, to process the data stored in the node.  Reduce involves taking the output from the nodes in the Map step to combine them into aggregate information.
